The data for defense expenditure and its spending as a percentage of total government spending is retrieved from the SIPRI Military Expenditure Database provided by Stockholm Military Peace Research Institute. The excel data file and pdfs are provided in the data section. Since we only plotted time slots for eight years, we figured it was more efficient to copy paste the values in rather than read in from the excel or pdf sheet. 

```python
import matplotlib.pyplot as plt

## Plot ROK Military Spending. Data from Sipri Milex Data
ROK_Military_Expenditure = [35298,36368, 37798, 39267, 40251, 40991, 43070, 46281]
ROK_Military_Expenditure_Yrs = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

plt.plot(ROK_Military_Expenditure_Yrs, ROK_Military_Expenditure, 'ro', 
         ROK_Military_Expenditure_Yrs, ROK_Military_Expenditure, 'k')
plt.xlabel('Year')
plt.ylabel('Defense Expenditure in $US M')
plt.title('ROK Military Spending')
plt.savefig('ROKSpending.png')
plt.show()
```

```python
ROK_Military_Expenditure = [12.7, 12.6, 12.8, 12.6, 12.6, 12.3, 12.2, 12.1]
ROK_Military_Expenditure_Yrs = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

plt.plot(ROK_Military_Expenditure_Yrs, ROK_Military_Expenditure, 'ro', 
         ROK_Military_Expenditure_Yrs, ROK_Military_Expenditure, 'k')
plt.xlabel('Year')
plt.ylabel('% of Govt Spending')
plt.title('ROK Defense Spending as Percentage of total Govt Spending')
plt.savefig('ROKPercentage.png')
plt.show()
```
